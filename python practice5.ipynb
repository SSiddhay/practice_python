```python
import matplotlib.pyplot as plt
x = np.arange(0,100)

y = x*2

z = x**2
plt.figure(figsize=(3,5))
plt.subplot(2,1,1)
plt.plot(x, y, color='blue', label='y = 2x')
plt.title('(x, y) Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

plt.figure(figsize=(3,5))
plt.subplot(2,1,2)
plt.plot(x, z, color='orange', label='z = x^2')
plt.title('(x, z) Plot')
plt.xlabel('x')
plt.ylabel('z')
plt.legend()
plt.show()
```

```python
diamond=pd.read_csv("diamonds.csv")
diamond
Diamond1 = diamond[(diamond['clarity'] == 'SI2') & (diamond['color'] == 'E')]
plt.figure(figsize=(10, 6))
sns.scatterplot(x=Diamond1['carat'],y=Diamond1['price'],hue=Diamond1['cut'] )
plt.xlabel('Carat',fontsize=20)
plt.ylabel('Price',fontsize=20)
plt.title('Scatter Plot of Carat vs. Price for SI2 Clarity and E Color Diamonds'
legend_labels = Diamond1['cut'].unique()
legend = plt.legend(legend_labels, title='Cut', loc='upper left')
plt.setp(legend.get_title(), fontsize='12')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
planet=pd.read_csv("planets.csv")
planet
plt.figure(figsize=(10, 6))
plt.subplot(2,1,1)
sns.histplot(planet['mass'],bins=30, color='skyblue')
plt.title('Distribution of Extrasolar Planet Masses', fontsize=16)
plt.xlabel('Mass', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
plt.figure(figsize=(10, 6))
plt.subplot(2,1,2)
sns.boxplot(x=planet['mass'], color='lightcoral')
plt.title('Boxplot of Extrasolar Planet Masses', fontsize=16)
plt.xlabel('Mass', fontsize=12)
plt.show()
```
