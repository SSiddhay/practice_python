```python
import numpy as np

# Creating a 2D array
array_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Flattening the 2D array to a 1D array
flattened_array = array_2d.flatten()

print("Original 2D Array:")
print(array_2d)
print("\nFlattened 1D Array:")
print(flattened_array)
```

```python
import numpy as np

# Creating a 2D array
array_2d = np.array([[1, 2, 3], [5, 4, 6]])

# Raveling the 2D array to a 1D array
raveled_array = np.ravel(array_2d)

print("Original 2D Array:")
print(array_2d)
print("\nRaveled 1D Array:")
print(raveled_array)
```

```python
def check_if_armstrong():
    number = input("Enter number: ")
    
    sum_of_cube = 0
    for i in range(len(number)):
        sum_of_cube = sum_of_cube + pow(int(number[i]), 3)
    print(sum_of_cube)
    
    if int(number) == sum_of_cube:
        print(f'{number} is armstrong')
    else:
        print(f'{number} isn\'t armstrong')
        
check_if_armstrong()
```

```python
def is_armstrong(num):
    # Count the number of digits
    num_str = str(num)
    num_digits = len(num_str)
    
    # Calculate the sum of each digit raised to the power of the number of digits
    armstrong_sum = sum(int(digit) ** num_digits for digit in num_str)
    print(armstrong_sum)
    print(num_digits)
    # Check if the sum is equal to the original number
    return armstrong_sum == num
    

# Example usage
number = int(input("Enter a number: "))
if is_armstrong(number):
    print(number, "is an Armstrong number.")
else:
    print(number, "is not an Armstrong number.")
```
